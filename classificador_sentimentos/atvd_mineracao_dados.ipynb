```python
# Criando as tuplas corretas de cada frase
basetreinamento = []
for frase in basedetreinos:
    # separando a frase em duas partes pela virgula
    parte = frase.split(', ')
    # se tiver duas partes
    if len(parte) == 2:
        # vai fazer uma tupla nova com as partes criadas separando os objetos pela virgula, e retirando o ponto final
        frase_formatada = (parte[0], parte[1].rstrip('.'))
        # adicionando as tuplas a um lista
        basetreinamento.append(frase_formatada)
basedeteste = []
for frase_teste in baseteste:
    # separando a frase em duas partes pela virgula
    parte_teste = frase_teste.split(', ')
    # se tiver duas partes
    if len(parte_teste) == 2:
        # vai fazer uma tupla nova com as partes criadas separando os objetos pela virgula, e retirando o ponto final
        frase_formatada_teste = (parte_teste[0], parte_teste[1].rstrip('.'))
        # adicionando as tuplas a uma lista
        basedeteste.append(frase_formatada_teste)
```

```python
stopwordsnltk = nltk.corpus.stopwords.words('portuguese')
stopwordsnltk.append('vou')
stopwordsnltk.append('tão')
stopwordsnltk.append('A')
stopwordsnltk.append('O')
```

```python
print(stopwordsnltk)
```

```python
def aplicastemmer(texto):
    stemmer = nltk.stem.RSLPStemmer()
    frasesstemming = []
    for (palavras, emocao) in texto:
        comstemming = [str(stemmer.stem(p)) for p in palavras.split()if p not in stopwordsnltk]
        frasesstemming.append((comstemming, emocao))
    return frasesstemming
```

```python
frases_stemming_treinamento = aplicastemmer(basetreinamento)
frases_stemming_teste = aplicastemmer(basedeteste)
print(frases_stemming_treinamento)
```

```python
todaspalavras_treinamento=[]
for (palavras1, emocao) in frases_stemming_treinamento:
     todaspalavras_treinamento.extend(palavras1)
todaspalavras_teste=[]
for (palavras2, emocao) in frases_stemming_teste:
     todaspalavras_teste.extend(palavras2)
```

```python
print(todaspalavras_teste)
```

```python
def buscafrequencia(palavras):
    palavras = nltk.FreqDist(palavras)
    return palavras
```

```python
frequencia_treinamento = buscafrequencia(todaspalavras_treinamento)
print(frequencia.most_common(50))
frequencia_teste = buscafrequencia(todaspalavras_teste)
```

```python
def buscapalavrasunicas(frequencia):
    freq = frequencia.keys()
    return freq
```

```python
palavrasunicas_treinamento = buscapalavrasunicas(frequencia_treinamento)
palavrasunicas_teste = buscapalavrasunicas(frequencia_teste)
print(palavrasunicas_treinamento)
```

```python
def extratorpalavras(documento):
    doc = set(documento)
    caracteristicas = {}
    for palavras in palavrasunicas_treinamento:
        caracteristicas['%s' % palavras] = (palavras in doc)
    return caracteristicas 
```

```python
basecompleta_treinamento = nltk.classify.apply_features(extratorpalavras, frases_stemming_treinamento)
basecompleta_teste = nltk.classify.apply_features(extratorpalavras, frases_stemming_teste)
```

```python
classificador = nltk.NaiveBayesClassifier.train(basecompleta_treinamento)
print(classificador.labels())
print(classificador.show_most_informative_features(150))
```

```python
print(nltk.classify.accuracy(classificador, basecompleta_teste))
```

```python
print(nltk.classify.accuracy(classificador, basecompleta_treinamento))
```
